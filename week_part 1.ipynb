```python
# send the GET request
data = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
soup.html.body.table
```

```python
for row in soup.find('table').find_all('tr'):
    cells = row.find_all('td')
    if(len(cells) > 0):
        print(cells[0].text,cells[2].find(text=True))
        postalCodeList.append(cells[0].text.strip('\n').rstrip('\n'))
        boroughList.append(cells[1].text.strip('\n').rstrip('\n'))
        neighborhoodList.append(cells[2].text)
        
```

```python
postalCodeList = []
boroughList = []
neighborhoodList = []
rows = soup.find('table').find_all('tr')
for row in rows[1:]:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    postalCodeList.append(cols[0])
    boroughList.append(cols[1])
    neighborhoodList.append(cols[2])
```

```python
toronto_df = pd.DataFrame({"PostalCode": postalCodeList,
                           "Borough": boroughList,
                           "Neighborhood": neighborhoodList})

toronto_df.head()
```

```python
toronto_df_dropna = toronto_df[toronto_df.Borough != "Not assigned"]
toronto_df_dropna.head()
```

```python
toronto_df_grouped = toronto_df_dropna.groupby(["PostalCode", "Borough"], as_index=False).agg(lambda x: ",".join(x))
toronto_df_grouped.head()
```

```python
for index, row in toronto_df_grouped.iterrows():
    if row["Neighborhood"] == "Not assigned":
        row["Neighborhood"] = row["Borough"]
        
toronto_df_grouped.head()
```
